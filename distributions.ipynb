```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exponential

from scipy.stats import expon

x = np.linspace(-3, 5, 1000)
pdf = expon.pdf(x)
plt.plot(x, pdf)
```

```python
# Different scales (1 /lambda)
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html

pdf = expon.pdf(x)
plt.plot(x, pdf)

pdf = expon.pdf(x, scale=2)
plt.plot(x, pdf)

pdf = expon.pdf(x, scale=3)
plt.plot(x, pdf)
```

```python
# Different locations

pdf = expon.pdf(x, loc=0)
plt.plot(x, pdf)

pdf = expon.pdf(x, loc=1)
plt.plot(x, pdf)
```

```python
# Normal
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

from scipy.stats import norm

pdf = norm.pdf(x)
plt.plot(x, pdf)
```

```python
# Different means

pdf = norm.pdf(x, loc=0)
plt.plot(x, pdf)

pdf = norm.pdf(x, loc=1)
plt.plot(x, pdf)
```

```python
# Different standard deviations

pdf = norm.pdf(x, scale=1)
plt.plot(x, pdf)

pdf = norm.pdf(x, scale=2)
plt.plot(x, pdf)
```

```python
# Lognormal
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lognorm.html

from scipy.stats import lognorm

x = np.linspace(0.01, 5, 1000)
pdf = lognorm.pdf(x, s=1)
plt.plot(x, pdf)
```

```python
# Different shapes
# Really weird

pdf = lognorm.pdf(x, s=1)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=0.2)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=0.5)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1.5)
plt.plot(x, pdf)
```

```python
# Different locations
# This just shifts it - boring

pdf = lognorm.pdf(x, s=1, loc=0)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, loc=1)
plt.plot(x, pdf)
```

```python
# Different scales

pdf = lognorm.pdf(x, s=1, scale=1)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=2)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=3)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=4)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=5)
plt.plot(x, pdf)
```

```python
# Beta
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html

from scipy.stats import beta

x = np.linspace(0, 1, 1000)
pdf = beta.pdf(x, a=2, b=5)
plt.plot(x, pdf)
```

```python
# Different a values

pdf = beta.pdf(x, a=1, b=5)
plt.plot(x, pdf)

pdf = beta.pdf(x, a=2, b=5)
plt.plot(x, pdf)

pdf = beta.pdf(x, a=3, b=5)
plt.plot(x, pdf)

pdf = beta.pdf(x, a=4, b=5)
plt.plot(x, pdf)
```

```python
# Different b values

pdf = beta.pdf(x, a=2, b=3)
plt.plot(x, pdf, color="red")

pdf = beta.pdf(x, a=2, b=4)
plt.plot(x, pdf, color="red")

pdf = beta.pdf(x, a=2, b=5)
plt.plot(x, pdf, color="red")


pdf = beta.pdf(x, a=1, b=3)
plt.plot(x, pdf, color="blue")

pdf = beta.pdf(x, a=1, b=4)
plt.plot(x, pdf, color="blue")

pdf = beta.pdf(x, a=1, b=5)
plt.plot(x, pdf, color="blue")
```
