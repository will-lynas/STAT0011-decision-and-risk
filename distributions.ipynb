```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exponential

from scipy.stats import expon

x = np.linspace(-3, 5, 1000)
pdf = expon.pdf(x)
plt.plot(x, pdf)
```

```python
# Different scales (1 /lambda)
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html

pdf = expon.pdf(x)
plt.plot(x, pdf)

pdf = expon.pdf(x, scale=2)
plt.plot(x, pdf)

pdf = expon.pdf(x, scale=3)
plt.plot(x, pdf)
```

```python
# Different locations

pdf = expon.pdf(x, loc=0)
plt.plot(x, pdf)

pdf = expon.pdf(x, loc=1)
plt.plot(x, pdf)
```

```python
# Normal
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

from scipy.stats import norm

pdf = norm.pdf(x)
plt.plot(x, pdf)
```

```python
# Different means

pdf = norm.pdf(x, loc=0)
plt.plot(x, pdf)

pdf = norm.pdf(x, loc=1)
plt.plot(x, pdf)
```

```python
# Different standard deviations

pdf = norm.pdf(x, scale=1)
plt.plot(x, pdf)

pdf = norm.pdf(x, scale=2)
plt.plot(x, pdf)
```
