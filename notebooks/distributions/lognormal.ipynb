```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lognorm.html

x = np.linspace(0.01, 5, 1000)
pdf = lognorm.pdf(x, s=1)
plt.plot(x, pdf)
```

```python
# Different shapes
# Really weird

pdf = lognorm.pdf(x, s=1)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=0.2)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=0.5)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1.5)
plt.plot(x, pdf)
```

```python
# Different locations
# This just shifts it - boring

pdf = lognorm.pdf(x, s=1, loc=0)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, loc=1)
plt.plot(x, pdf)
```

```python
# Different scales

pdf = lognorm.pdf(x, s=1, scale=1)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=2)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=3)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=4)
plt.plot(x, pdf)

pdf = lognorm.pdf(x, s=1, scale=5)
plt.plot(x, pdf)
```
