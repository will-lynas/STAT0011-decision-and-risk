## Exponential
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html

<img src="../../images/expon.png" width="500">

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import expon

x = np.linspace(-3, 5, 1000)
pdf = expon.pdf(x)
plt.plot(x, pdf)
```

```python
# Different scales (1/lambda)

pdf = expon.pdf(x)
plt.plot(x, pdf)

pdf = expon.pdf(x, scale=2)
plt.plot(x, pdf)

pdf = expon.pdf(x, scale=3)
plt.plot(x, pdf)
```

```python
# Different locations

pdf = expon.pdf(x, loc=0)
plt.plot(x, pdf)

pdf = expon.pdf(x, loc=1)
plt.plot(x, pdf)
```
