```python
import numpy as np  
import matplotlib.pyplot as plt
from scipy.stats import genpareto

x = np.linspace(0, 10, 1000)
pdf = genpareto.pdf(x, c=1)
plt.plot(x, pdf)
```

```python
# c=0 is exponential
pdf = genpareto.pdf(x, c=0)
plt.plot(x, pdf, label="c=0")

pdf = genpareto.pdf(x, c=1)
plt.plot(x, pdf, label="c=1")

pdf = genpareto.pdf(x, c=3)
plt.plot(x, pdf, label="c=10")

plt.legend()
```

```python

# Negative values

pdf = genpareto.pdf(x, c=-0.2)
plt.plot(x, pdf, label="c=-0.2")

pdf = genpareto.pdf(x, c=-0.4)
plt.plot(x, pdf, label="c=-0.4")

pdf = genpareto.pdf(x, c=-0.5)
plt.plot(x, pdf, label="c=-0.5")

pdf = genpareto.pdf(x, c=-1)
plt.plot(x, pdf, label="c=-1")

plt.legend()
```
