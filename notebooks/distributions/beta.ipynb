```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html

x = np.linspace(0, 1, 1000)
pdf = beta.pdf(x, a=2, b=5)
plt.plot(x, pdf)
```

```python
# Different a values

pdf = beta.pdf(x, a=1, b=5)
plt.plot(x, pdf)

pdf = beta.pdf(x, a=2, b=5)
plt.plot(x, pdf)

pdf = beta.pdf(x, a=3, b=5)
plt.plot(x, pdf)

pdf = beta.pdf(x, a=4, b=5)
plt.plot(x, pdf)
```

```python
# Different b values

pdf = beta.pdf(x, a=2, b=3)
plt.plot(x, pdf, color="red")

pdf = beta.pdf(x, a=2, b=4)
plt.plot(x, pdf, color="red")

pdf = beta.pdf(x, a=2, b=5)
plt.plot(x, pdf, color="red")


pdf = beta.pdf(x, a=1, b=3)
plt.plot(x, pdf, color="blue")

pdf = beta.pdf(x, a=1, b=4)
plt.plot(x, pdf, color="blue")

pdf = beta.pdf(x, a=1, b=5)
plt.plot(x, pdf, color="blue")
```
