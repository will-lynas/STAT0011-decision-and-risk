```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pct_change = np.linspace(-50, 50, 100)

initial_price = 100
final_price = initial_price * (1 + pct_change / 100)

simple_return = (final_price - initial_price) / initial_price
log_return = np.log(final_price / initial_price)

diff = simple_return - log_return

plt.figure(figsize=(10, 5))
plt.plot(pct_change, simple_return, label='Simple Return')
plt.plot(pct_change, log_return, label='Log Return')
plt.legend()
plt.show()
```
